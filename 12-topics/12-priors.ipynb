# Dirichlet distributions

Visualize effect of changing concentration parameter $\alpha$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plot(alpha):
    r = np.random.dirichlet(alpha,1000000)
    plt.hexbin(r[:,0], r[:,1], bins='log', cmap='plasma')
    #plt.hist2d(r[:,0], r[:,1], cmap='plasma')
    plt.title(f'Alpha = {alpha}')
    plt.xlabel('P(A)')
    plt.ylabel('P(B)')
```

```python
plot([1,1,1])
```

```python
plot([10,10,10])
```

```python
plot([0.1,0.1,0.1])
```

```python
plot([0.1,1,1])
```

```python
plot([0.1,1,10])
```
